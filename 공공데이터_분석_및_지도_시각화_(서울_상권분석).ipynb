### 지도 시각화 분석

```python
!pip install folium
```

```python
import folium
```

```python
# 1) 표시될 기본 지도 그리기
map1 = folium.Map(location=[37.542539, 127.133575])
```

```python
# 2) 해당 매장위치 마킹
folium.Marker(location=[37.542539, 127.133575]).add_to(map1)
```

```python
map1
```

```python
# 위도, 경도 평균 구해서 지도 그리기 (지도의 초점을 맞추기 위해)
l = shop_coffee_idy[["위도"]].mean()
m = shop_coffee_idy[["경도"]].mean()

map2 = folium.Map(location=[l, m], zoom_start=10)
```

```python
# 이디야 매장위치 마킹하기

for idx in shop_coffee_idy.index:
  a = shop_coffee_idy.loc[idx, "위도"]
  b = shop_coffee_idy.loc[idx, "경도"]
  folium.Marker(location=[a, b]).add_to(map2)
```

```python
map2
```

```python
# 이디야 매장위치 마킹하기 개선된 방법 cluster 사용하기
from folium.plugins import MarkerCluster
```

```python
map3 = folium.Map(location=[l, m], zoom_start=10)

cluster = MarkerCluster()
c_map = cluster.add_to(map3)
```

```python
for idx in shop_coffee_idy.index:
  a = shop_coffee_idy.loc[idx, "위도"]
  b = shop_coffee_idy.loc[idx, "경도"]
  c = shop_coffee_idy.loc[idx, "상호명"]
  d = shop_coffee_idy.loc[idx, "시군구명"]
  folium.Marker(location=[a, b],
                popup=c,
                tooltip=d,
                icon=folium.Icon("green")).add_to(c_map)
```
